```python
# prompt: generate a data set for a set of employee salaries and bell graph
# generate it  in graph pattern

import matplotlib.pyplot as plt
import numpy as np

# Generate sample salary data (replace with your actual data)
np.random.seed(0)  # for reproducibility
salaries = np.random.normal(loc=60000, scale=15000, size=100)

# Create the bell curve (histogram)
plt.hist(salaries, bins=20, density=True, alpha=0.6, color='g')

# Add a smooth bell curve (Gaussian distribution)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = 1/(15000 * np.sqrt(2 * np.pi)) * np.exp( - (x - 60000)**2 / (2 * 15000**2))
plt.plot(x, p, 'k', linewidth=2)

# Customize the plot
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.title("Employee Salaries Distribution")
plt.grid(True)

# Display the plot
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup

# Step 1: URL to scrape
url = 'https://www.geeksforgeeks.org/introduction-of-object-oriented-programming/'

# Step 2: Send a GET request
response = requests.get(url)

# Step 3: Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Step 4: Extract visible text from <p>, <h1>-<h3>, etc.
content = []

# Add the title of the page
page_title = soup.title.string if soup.title else 'No Title Found'
content.append(f"Page Title: {page_title}\n")

# Extract paragraphs
for para in soup.find_all('p'):
    text = para.get_text(strip=True)
    if text:
        content.append(text)

# Extract headings
for tag in ['h1', 'h2', 'h3']:
    for heading in soup.find_all(tag):
        text = heading.get_text(strip=True)
        if text:
            content.append(f"{tag.upper()}: {text}")

# Step 5: Save content to a text file
with open("scraped_content.txt", "w", encoding="utf-8") as file:
    for line in content:
        file.write(line + "\n")

print("✅ Content downloaded and saved to 'scraped_content.txt'")
```
